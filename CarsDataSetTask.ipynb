```python
#import Libraries (numpy & Pandas)
import numpy as np 
import pandas as pd
```

```python
#read Data
df = pd.read_csv(r"used_car_prices.csv")
```

```python
#print the first 8 row
df.head(8)
```

```python
#print the shape Like (There are 84548 rows and 22 columns)
print(f"There are {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# try seeing some information about the data
df.info()
```

```python
# show the number of duplicates here
df.duplicated().sum()
```

```python
# show number of nulls
df.isnull().sum()
```

```python
# updated the column names Like(To Start with Upper Litter)
df.rename(columns={"web-scraper-order":"Web Scraper Order"})
```

```python
# Split Car Full Data into Brand, Model, And Year of The Model
df["Car Model"] = df["Car Model"].apply(lambda x : x.split())
df["Brand"] = df["Car Model"].apply(lambda x : x[0])
df["Model"] = df["Car Model"].apply(lambda x : ' '.join(x[1:-1]))
df["Year of The Model"] = df["Car Model"].apply(lambda x : x[-1])
df.drop(columns='Car Model',inplace=True)
```

```python
# Convert Average price, Minimum price, Maximum price to Numaarical Values
df["Average price"] = df["Average price"].str.strip(" EGP").str.replace(",","").astype(float)
df["Minimum price"] = df["Minimum price"].str.strip(" EGP").str.replace(",","").astype(float)
df["Maximum price"] = df["Maximum price"].str.strip(" EGP").str.replace(",","").astype(float)
```

```python
# Split Month/Year Column to Year and Month 
df["Month"] = df["Month/Year"].apply( lambda x: str(x).split('-')[-1])
df["Year"] = df["Month/Year"].apply( lambda x: str(x).split('-')[0])
df.drop(columns= "Month/Year",inplace=True)
```

```python
# let's start with (Average price ,Minimum price ,Maximum price ) , see the number of nulls in them , replace these nulls with the ( mean , median and mode) of these columns
#and in the end check that there are not nulls in data 
print(df["Average price"].isnull().sum(),end=", ")
print(df["Minimum price"].isnull().sum(),end=", ")
print(df["Maximum price"].isnull().sum())

df["Average price"].fillna(df["Average price"].mean(),inplace=True)
df["Minimum price"].fillna(df["Minimum price"].mode()[0],inplace=True)
df["Maximum price"].fillna(df["Maximum price"].median(),inplace=True)

print(df["Average price"].isnull().sum(),end=", ")
print(df["Minimum price"].isnull().sum(),end=", ")
print(df["Maximum price"].isnull().sum(),end=" ")
```
